```python
sales = float(input("Enter sales ($): "))
rate = float(input("Enter tax rate (%): "))
print(f"Sales: ${sales}")

# Cost of Goods Sold
cog = 0.40
cog_sold = sales * cog
print(f"Cost of Goods Sold: ${cog_sold}")

# Gross Profit
gross_profit = sales - (sales * cog)
print(f"Gross Profit: ${gross_profit}")

# Taxes
taxes = (rate / 100) * gross_profit
print(f"Taxes: ${taxes}")

# Net Incomex
net_income = gross_profit - taxes
print(f"Net Income: ${net_income}")
```
