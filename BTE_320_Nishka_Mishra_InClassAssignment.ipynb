Income Statment

```python
# Inputs: sales, cogs_pct and tax rate
sales = input("Enter sales ($): ")
sales = float(sales)

cogs_pct = input("Enter Cost of Goods sold (%): ")
cogs_pct = float(cogs_pct)

rate = input("Enter tax rate (%): ")
rate = float(rate)

# Processes: calculate actual cost of goods sold, taxes and net income
cogs = sales * cogs_pct / 100
grossProfit = sales - cogs
taxes = grossProfit * (rate / 100)
netIncome = grossProfit - taxes

#Outputs: display cogs and grossProfit
print(f"Cost of Goods sold: ${cogs}\nGross Profit: ${grossProfit}\nTaxes: ${taxes}\nNet Income: ${netIncome}")
#print(f"Gross Profit: ${grossProfit}")
```

Calculating EOQ

```python
#Inputs: demand, reorder cost, holding cost
demand = float(input("Enter the projected demand (units/year): "))
reorderCost = float(input("Enter the reorder cost ($/order): "))
holdingCost = float(input("Enter the holding cost ($/year/unit): "))

#Process: calculate EOQ
eoq = (2 * demand * reorderCost / holdingCost) ** 0.5

#Output: display EOQ value
print(f"Economic Order Quantity: {eoq}")
```

Rock Paper Scissors

```python
import random

def play_rock_paper_scissors():
    """
    Implements a Rock, Paper, Scissors game using a while loop.
    The game continues as long as the user wants to play.
    """
    # 1. Initial check to start the game
    start_game = input("Do you want to play Rock, Paper, Scissors? (yes/no): ").strip().lower()

    if start_game != 'yes':
        print("Maybe next time! Goodbye! üëã")
        return # Exit the function if the user doesn't want to play

    print("\n--- Game Starting! ---\n")

    # 2. Main game loop
    while start_game == 'yes':
        # List of valid choices for the computer
        choices = ['rock', 'paper', 'scissors']

        # Computer makes a random choice
        computer_choice = random.choice(choices)

        # Get the user's choice
        user_choice = input("Enter your choice (rock, paper, scissors) or 'quit' to stop: ").strip().lower()

        # Check if the user wants to quit
        if user_choice == 'quit':
            print("\nThanks for playing! Final score is not tracked, but you had fun! üòâ")
            break # Exit the while loop

        # Input validation
        if user_choice not in choices:
            print(f"Invalid choice: '{user_choice}'. Please enter 'rock', 'paper', or 'scissors'.")
            continue # Go back to the start of the loop to ask again

        print(f"\nYou chose: **{user_choice.capitalize()}**")
        print(f"Computer chose: **{computer_choice.capitalize()}**\n")

        # 3. Determine the winner
        if user_choice == computer_choice:
            print("It's a tie! ü§ù")
        elif (user_choice == 'rock' and computer_choice == 'scissors') or \
             (user_choice == 'paper' and computer_choice == 'rock') or \
             (user_choice == 'scissors' and computer_choice == 'paper'):
            print("üéâ **You win!** Congratulations! üéâ")
        else:
            print("ü§ñ **Computer wins!** Better luck next time. üòî")

        # 4. Ask the user if they want to play another round
        play_again = input("\nPlay another round? (yes/no): ").strip().lower()
        print("----------------------------")

        if play_again != 'yes':
            print("Thanks for playing! Goodbye! üëã")
            break # Exit the while loop

# Execute the game function
if __name__ == "__main__":
    play_rock_paper_scissors()
```

Quality Control Manager

```python
def getValues(n: int) -> list:
  '''
  Prompts the user for inputs, adds them to a list

  Input:
  - Number of items in the list

  Output:
  - A list
  '''
  items = []

  for i in range(n):
    try:
      item = float(input('Enter a weight value: '))
    except Exception as e:
      item = float('nan')

    items.append(item)
  print(items)
  return items

def calcAverage(alist: list) -> float:
  '''
  Input:
  - a list of weights

  Output:
  - the average weight
  '''
  try:
    return sum(alist[2:]) / len(alist[2:])
  except Exception as e:
    print(e)
    return float('nan')

def main():
  n = int(input('How many weights? '))
  if n < 0:
    raise ValueError('Number of items must be positive!')

  L = getValues(n)
  avg = calcAverage(L)
  print(f'Average weight: {avg} lbs.')

main()
```

ROI Calculator

```python
def roi(cost: float, revenue: float) -> float:
    assert cost > 0, "Cost must be positive"
    assert isinstance(cost, (int, float)), "Cost must be numerical"
    assert isinstance(revenue, (int, float)), "Revenue must be numerical"

    return (revenue - cost) / cost

if __name__ == "__main__":
    c = float(input('Enter cost:'))
    r = float(input('Enter revenue:'))
    ret = roi(c, r)
    print(ret)
```

Loan Payment Calculator

```python
def loan_payment(principal: float, annual_rate: float, months: int) -> float:
    assert isinstance(principal, (int, float)), "principal must be numeric"
    assert isinstance(annual_rate, (int, float)), "annual_rate must be numeric"
    assert isinstance(months, int), "months must be an integer"

    if principal <= 0:
        raise ValueError("Principal must be > 0.")
    if annual_rate < 0:
        raise ValueError("Annual rate cannot be negative.")
    if months <= 0:
        raise ValueError("Number of months must be > 0.")

    r = annual_rate / 12.0
    n = months
    P = float(principal)

    if r == 0.0:
        return P / n
    else:
        return r * P / (1.0 - (1.0 + r) ** (-n))

if __name__ == "__main__":
    try:
        P = float(input("Enter principal: "))
        apr = float(input("Enter annual interest rate as a decimal: "))
        n = int(input("Enter number of months: "))
        payment = loan_payment(P, apr, n)
        print(f"Monthly payment: ${payment:.2f}")
    except Exception as e:
        print(e)
```

Class Greeting

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def __str__(self):
    return "Instance of class Greeting"

  def display(self, reverse=True):
    if reverse:
      print(f'{self.last}, {self.first}')
    else:
      print(self.first)

name = Greeting('John', 'Doe')
print(name)
name.display()
name.display(reverse=False)
```

Student Class

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.__number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newname):
    self.__name = newname

  def __len__(self):
    return len(self.__courses)

  def enroll(self, course):
    if course in self.__courses:
      print('You have already enrolled in', course)
    else:
      self.__courses.append(course)

s = Student('Bob', 12345)
s.get_name()
s.get_courses()
s.set_name('Bob Andrews')
s.get_name()
s.enroll('ABC123')
s.enroll('ABC123')
s.enroll('ABC567')
len(s)
s.get_courses()
```

Classes Point and Rectangle

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** .5

class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(4,0)
    self.p3 = Point(4,4)
    self.p4 = Point(0,4)
    self.width = self.p1.distance(self.p2)
    self.height = self.p2.distance(self.p3)

  def area(self):
    return self.width * self.height
```
